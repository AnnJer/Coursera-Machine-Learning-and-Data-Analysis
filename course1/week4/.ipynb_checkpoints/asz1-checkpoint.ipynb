```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

### Step 1: Interesting distribution - Erlang

```python
rv = sts.erlang(8,2 )
```

#### Сгенерируйте из него выборку объёма 1000

```python
sample = rv.rvs(1000)
```

1. Постройте гистограмму выборки 
2. Нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины

```python
# 1
x = np.linspace(min(sample),max(sample),1000)
plt.plot(x, rv.pdf(x), label='theoretical PDF')
# 2
plt.hist(sample, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.legend()
```

### Step 2
1. Для нескольких значений n (например, 5, 10, 50) сгенерируйте 1000 выборок объёма n.
2. Постройте гистограммы распределений их выборочных средних.
3. Посчитайте значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних.

```python
# вычисление параметров распределения
k = 16
theta = 1.05
rv = sts.erlang(k, theta)
mu = k * theta
sigma2 = k * np.power(theta, 2)
for N in [5, 10, 50]:
    sample = [np.mean(rv.rvs(N)) for _ in xrange(1000)]
    x = np.linspace(min(sample), max(sample), 1000)
    norm_rv = sts.norm(mu, sigma2/float(N))
    plt.plot(x, norm_rv.pdf(x), label="N=%s: %s, %s" % (N, mu, sigma2/float(N)))
    plt.ylabel('number of samples')
    plt.xlabel('$x$')
    plt.hist(sample, normed=True)
    plt.legend(loc='upper left')
plt.show()
```

```python
# вычисление параметров распределения
k = 16
theta = 1.05
rv = sts.erlang(k, theta)
mu = k * theta
sigma2 = k * np.power(theta, 2)
for N in [5, 10, 50]:
    sample = [np.mean(rv.rvs(N)) for _ in xrange(1000)]
    x = np.linspace(min(sample), max(sample), 1000)
    norm_rv = sts.norm(mu, sigma2/float(N))
    plt.plot(x, norm_rv.pdf(x), label="N=%s: %s, %s" % (N, mu, sigma2/float(N)))
    plt.ylabel('number of samples')
    plt.xlabel('$x$')
    plt.hist(sample, normed=True)
    plt.legend(loc='upper left')
    plt.show()
```

### Вывод

При увлечении выборки значения приближаются к матожиданию (собственно, ЦПТ).
Хотя распределение и стремится к нормальному, результат не идеален.
